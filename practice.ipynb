```python
from pathlib import Path
import json
import pandas as pd

def load_json_files_separately(data_dir="api_data_files"):
    """
    Reads all .json files in the folder and returns a dictionary
    mapping each filename -> its corresponding DataFrame.
    """
    data_dir = Path(data_dir)
    dfs = {}

    for file in sorted(data_dir.glob("*.json")):
        with open(file, "r", encoding="utf-8") as f:
            data = json.load(f)               # load list of dicts
        df = pd.json_normalize(data)          # flatten to DataFrame
        dfs[file.stem] = df                   # store by filename (no .json)
        print(f"Loaded {file.name}")

    return dfs

dataframes = load_json_files_separately("api_data_files")
```

```python
import re
import pandas as pd
from fuzzywuzzy import process, fuzz

# Strip trailing name suffixes like "Jr.", "Sr", "III" from LAST NAME specifically
_SUFFIX_TRAILING_RE = re.compile(r"(?:,?\s*(?:jr|sr|ii|iii|iv|v))\.?$", re.IGNORECASE)

def _norm_name(first: str, last: str) -> str:
    """Normalize full name: lowercase, trim, remove suffixes from last name, collapse whitespace."""
    f = (str(first) if pd.notna(first) else "").strip().lower()
    l = (str(last)  if pd.notna(last)  else "").strip().lower()

    # remove a trailing suffix if it's attached to the last name, e.g., "greene jr.", "greene, Jr", "greene jr"
    l = _SUFFIX_TRAILING_RE.sub("", l).strip()

    # build normalized full name and collapse any extra spaces
    full = " ".join([p for p in [f, l] if p])
    full = " ".join(full.split())
    return full

def concat_and_fuzzy_merge(international_df: pd.DataFrame,
                           nba_df: pd.DataFrame,
                           player_df: pd.DataFrame,
                           threshold: int = 85) -> pd.DataFrame:
    combined_df = pd.concat([international_df, nba_df], ignore_index=True, sort=False)

    combined_df["full_name_norm"] = [
        _norm_name(f, l) for f, l in zip(combined_df["first_name"], combined_df["last_name"])
    ]
    player_df = player_df.copy()
    player_df["full_name_norm"] = [
        _norm_name(f, l) for f, l in zip(player_df["first_name"], player_df["last_name"])
    ]

    # exact match, then fuzzy fallback
    exact_map = dict(zip(player_df["full_name_norm"], player_df["full_name_norm"]))
    combined_df["matched_name"] = combined_df["full_name_norm"].map(exact_map)
    combined_df["match_score"] = combined_df["matched_name"].notna().astype(int) * 100

    search_space = player_df["full_name_norm"].dropna().unique().tolist()

    def _fuzzy_find(q: str):
        if not q or pd.isna(q):
            return (None, 0)
        match, score = process.extractOne(q, search_space, scorer=fuzz.token_sort_ratio) or (None, 0)
        return (match, score) if score >= threshold else (None, score)

    need_fuzzy = combined_df["matched_name"].isna()
    if need_fuzzy.any():
        fuzzy_results = combined_df.loc[need_fuzzy, "full_name_norm"].apply(_fuzzy_find)
        combined_df.loc[need_fuzzy, "matched_name"] = [r[0] for r in fuzzy_results]
        combined_df.loc[need_fuzzy, "match_score"] = [r[1] for r in fuzzy_results]

    out = combined_df.merge(
        player_df[["full_name_norm", "birth_date"]].rename(columns={"full_name_norm": "matched_name"}),
        on="matched_name", how="left"
    ).drop(columns=["full_name_norm"])

    return out

combined_df = concat_and_fuzzy_merge(international_df, nba_df, player_df, threshold=90)
combined_df[combined_df["last_name"]== "haslem"]
```

```python
import re
import pandas as pd
import numpy as np

_SUFFIX_TRAILING_RE = re.compile(r"(?:,?\s*(?:jr|sr|ii|iii|iv|v))\.?$", re.IGNORECASE)

def _norm_first_last(first: str, last: str) -> str:
    f = (str(first) if pd.notna(first) else "").strip().lower()
    l = (str(last)  if pd.notna(last)  else "").strip().lower()
    l = _SUFFIX_TRAILING_RE.sub("", l).strip()
    full = " ".join([p for p in [f, l] if p])
    return " ".join(full.split())

def merge_player_seasons(combined_df: pd.DataFrame) -> pd.DataFrame:
    df = combined_df.copy()

    # Ensure matched_name
    if "matched_name" not in df.columns:
        df["matched_name"] = np.nan
    df["matched_name"] = df["matched_name"].astype(str).str.strip()
    need_fill = df["matched_name"].eq("") | df["matched_name"].isna()
    if need_fill.any():
        df.loc[need_fill, "matched_name"] = [
            _norm_first_last(f, l) for f, l in zip(df.get("first_name", ""), df.get("last_name", ""))
        ]
    df["matched_name"] = df["matched_name"].astype(str).str.strip().str.lower()

    # Columns
    all_cols = [
        "matched_name","first_name","last_name","season","season_type","league","team",
        "games","starts","minutes","points","two_points_made","two_points_attempted",
        "three_points_made","three_points_attempted","free_throws_made","free_throws_attempted",
        "blocked_shot_attempts","offensive_rebounds","defensive_rebounds","assists","screen_assists",
        "turnovers","steals","deflections","loose_balls_recovered","blocked_shots","personal_fouls",
        "personal_fouls_drawn","offensive_fouls","charges_drawn","technical_fouls","flagrant_fouls",
        "ejections","points_off_turnovers","points_in_paint","second_chance_points","fast_break_points",
        "possessions","estimated_possessions","team_possessions","usage_percentage",
        "true_shooting_percentage","three_point_attempt_rate","free_throw_rate",
        "offensive_rebounding_percentage","defensive_rebounding_percentage","total_rebounding_percentage",
        "assist_percentage","steal_percentage","block_percentage","turnover_percentage",
        "internal_box_plus_minus","field_goals_attempted","age"
    ]
    for c in all_cols:
        if c not in df.columns:
            df[c] = np.nan

    # Types
    df["season"] = pd.to_numeric(df["season"], errors="coerce")
    df["age"] = pd.to_numeric(df["age"], errors="coerce")
    text_like = {"matched_name","first_name","last_name","season_type","league","team"}
    numeric_like = set(all_cols) - text_like
    for c in numeric_like:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # >>> Keys now include AGE <<<
    keys = ["matched_name","age","season"]

    # Aggregations
    sum_cols = [
        "games","starts","minutes","points",
        "two_points_made","two_points_attempted",
        "three_points_made","three_points_attempted",
        "free_throws_made","free_throws_attempted",
        "blocked_shot_attempts","offensive_rebounds","defensive_rebounds",
        "assists","screen_assists","turnovers","steals","deflections",
        "loose_balls_recovered","blocked_shots","personal_fouls","personal_fouls_drawn",
        "offensive_fouls","charges_drawn","technical_fouls","flagrant_fouls","ejections",
        "points_off_turnovers","points_in_paint","second_chance_points","fast_break_points",
        "possessions","estimated_possessions","team_possessions",
    ]
    agg_sum = {c: "sum" for c in sum_cols if c in df.columns}
    grouped = df.groupby(keys, dropna=False).agg(agg_sum).reset_index()

    # season_type
    def choose_season_type(g):
        vals = g["season_type"].dropna().astype(str)
        if len(vals) == 0:
            return np.nan
        if "Full Season" in set(vals):
            return "Full Season"
        return vals.mode().iat[0]

    season_type_map = (df.groupby(keys, dropna=False)
                         .apply(choose_season_type)
                         .rename("season_type")
                         .reset_index())
    grouped = grouped.merge(season_type_map, on=keys, how="left")

    # league/team summaries
    def summarize_text(g, col):
        vals = g[col].dropna().astype(str).unique().tolist()
        if not vals:
            return np.nan
        return " + ".join(sorted(vals)) if len(vals) > 1 else vals[0]

    meta = (df.groupby(keys, dropna=False)
              .apply(lambda g: pd.Series({
                  "league": summarize_text(g, "league"),
                  "team": summarize_text(g, "team")
              }))
              .reset_index())
    grouped = grouped.merge(meta, on=keys, how="left")

    # Minutes-weighted averages for rate stats
    rate_cols = [
        "usage_percentage",
        "offensive_rebounding_percentage","defensive_rebounding_percentage","total_rebounding_percentage",
        "assist_percentage","steal_percentage","block_percentage","turnover_percentage",
        "internal_box_plus_minus",
    ]
    def wavg(g, col):
        m = g["minutes"].fillna(0)
        x = pd.to_numeric(g[col], errors="coerce")
        wsum = (x*m).sum()
        msum = m.sum()
        return np.nan if msum == 0 else wsum / msum

    w_means = (df[keys + ["minutes"] + rate_cols]
               .groupby(keys, dropna=False)
               .apply(lambda g: pd.Series({c: wavg(g, c) for c in rate_cols}))
               .reset_index())
    grouped = grouped.merge(w_means, on=keys, how="left")

    # --- Derived shooting/rates ---
    grouped["field_goals_attempted"] = (
        grouped["two_points_attempted"].fillna(0) + grouped["three_points_attempted"].fillna(0)
    )
    grouped["three_point_attempt_rate"] = np.where(
        grouped["field_goals_attempted"] > 0,
        grouped["three_points_attempted"] / grouped["field_goals_attempted"],
        np.nan
    )
    grouped["free_throw_rate"] = np.where(
        grouped["field_goals_attempted"] > 0,
        grouped["free_throws_attempted"] / grouped["field_goals_attempted"],
        np.nan
    )
    denom = (grouped["field_goals_attempted"] + 0.44 * grouped["free_throws_attempted"])
    grouped["true_shooting_percentage"] = np.where(
        denom > 0,
        grouped["points"] / (2 * denom),
        np.nan
    )
    grouped["ppg"] = np.where(grouped["games"] > 0, grouped["points"] / grouped["games"], np.nan)
    grouped["apg"] = np.where(grouped["games"] > 0, grouped["assists"] / grouped["games"], np.nan)

    total_fgm = grouped["two_points_made"].fillna(0) + grouped["three_points_made"].fillna(0)
    grouped["efg"] = np.where(
        grouped["field_goals_attempted"] > 0,
        (total_fgm + 0.5 * grouped["three_points_made"]) / grouped["field_goals_attempted"],
        np.nan
    )

    grouped["four_factors_score"] = (
        0.4  * grouped["efg"]
        - 0.25 * grouped["turnover_percentage"]
        + 0.2  * grouped["offensive_rebounding_percentage"]
        + 0.15 * grouped["free_throw_rate"]
    )

    # If you no longer want first/last in the output:
    # grouped = grouped.drop(columns=["first_name","last_name"], errors="ignore")

    return grouped

# usage
merged_df = merge_player_seasons(combined_df)
merged_df

```

```python
# We want to assess the predictiveness of EURO success on NBA success
# in other words, how well does european production translate?

import numpy as np
import pandas as pd

# ============================================================
# GOAL:
# Assess how well European (non-NBA) production translates to NBA success
# ============================================================

# Expect merged_df to have at least: matched_name, season, league, games,
# points, offensive_rebounds, defensive_rebounds, assists

# ============================================================
# 1. Compute NBA stickiness (4+ seasons)
# ============================================================

# Filter NBA players
nba_players = merged_df[merged_df["league"] == "NBA"].copy()

# Count unique NBA seasons per player (use matched_name)
stickiness = (
    nba_players.groupby(["matched_name"], as_index=False)["season"]
    .nunique()
    .rename(columns={"season": "nba_seasons"})
)

# Merge stickiness back into master dataset
all_players = merged_df.merge(stickiness, on=["matched_name"], how="left")

# Fill missing (non-NBA players) and flag stickiness
all_players["nba_seasons"] = all_players["nba_seasons"].fillna(0).astype(int)
all_players["nba_sticky"] = all_players["nba_seasons"] >= 3

# ============================================================
# 2. Compute NBA Production Scores
# ============================================================

nba_df = all_players.query("league == 'NBA'").copy()

nba_df["production_score"] = (
    (nba_df["points"]
     + nba_df["offensive_rebounds"]
     + nba_df["defensive_rebounds"]
     + nba_df["assists"])
    / nba_df["games"].replace(0, np.nan)
)

# Compute season-level mean and std for production
nba_stats = (
    nba_df.groupby("season")["production_score"]
    .agg(["mean", "std"])
    .reset_index()
    .rename(columns={"mean": "avg_production", "std": "std_production"})
)

# Merge NBA stats back to player-season level
nba_df = nba_df.merge(nba_stats, on="season", how="left")

# Compute z-scores (guard against std=0)
nba_df["production_z"] = np.where(
    (nba_df["std_production"] > 0) & nba_df["std_production"].notna(),
    (nba_df["production_score"] - nba_df["avg_production"]) / nba_df["std_production"],
    np.nan
)

# Flag production success (boolean)
nba_df["production_success"] = nba_df["production_z"] > 1.96

# Keep only relevant NBA success columns for merge
nba_success_cols = [
    "matched_name", "season",
    "production_score", "production_z", "production_success"
]

# ============================================================
# 3. Merge NBA success metrics back into the unified player table
# ============================================================

nba_and_int_players = (
    all_players
    .merge(
        nba_df[nba_success_cols],
        on=["matched_name", "season"],
        how="left",
        validate="1:1"  # sanity check, ensures no duplication
    )
)

# ============================================================
# 4. Final sanity cleanup
# ============================================================

# Prevent leftover merge duplicates (nba_seasons_x, _y)
nba_and_int_players = nba_and_int_players.loc[:, ~nba_and_int_players.columns.duplicated()]

# Sort for readability (use matched_name primary)
nba_and_int_players = nba_and_int_players.sort_values(
    by=["matched_name", "season"]
).reset_index(drop=True)

# ============================================================
# 5. Output clean dataset (successful NBA production seasons)
# ============================================================

nba_and_int_players[nba_and_int_players["production_success"] == True]
```

```python
# we are going to run 2 logistic regressions 
# (This block currently uses RandomForest as in your snippet; swap to LogisticRegression if desired.)

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

nba_and_int_players["is_euroleague"] = (nba_and_int_players["league"].str.contains("EuroLeague")).astype(int)
nba_and_int_players["is_acb"] = (nba_and_int_players["league"].str.contains("ACB")).astype(int)
nba_and_int_players["italy"] = (nba_and_int_players["league"].str.contains("Italy")).astype(int)
nba_and_int_players["eurocup"] = (nba_and_int_players["league"].str.contains("EuroCup")).astype(int)



# Outcome variable 1 = NBA Sticky
# For each player, identify first NBA season
first_nba = (
    nba_and_int_players[nba_and_int_players["league"] == "NBA"]
    .groupby(["matched_name"], as_index=False)["season"]
    .min()
    .rename(columns={"season": "first_nba_season"})
)

# Merge and filter to only seasons before NBA debut (non-NBA seasons only)
pre_nba = (
    nba_and_int_players
    .merge(first_nba, on=["matched_name"], how="left")
    .query("league != 'NBA' and season < first_nba_season")
)

pre_nba_agg = (
    pre_nba.groupby(["matched_name"], as_index=False)
    .agg({
        "ppg": ["mean", "max"],
        "apg": ["mean", "max"],
        "usage_percentage": "mean",
        "three_point_attempt_rate": "mean",
        "four_factors_score": "mean",
        "true_shooting_percentage": "mean",
        "steals": "mean",
        "age": ["min", "max"],
        "is_euroleague": "mean",   # % of seasons in EuroLeague
        "is_acb": "mean",
        "italy": "mean",
        "eurocup": "mean",
        "internal_box_plus_minus": "mean",
        "season": "count"
    })
)

# Flatten multi-level columns
pre_nba_agg.columns = ["_".join(col).rstrip("_") if isinstance(col, tuple) else col
                       for col in pre_nba_agg.columns]
pre_nba_agg = pre_nba_agg.rename(columns={"season_count": "num_pre_nba_seasons"})

# Bring in outcome label at player level
player_level = pre_nba_agg.merge(
    nba_and_int_players[["matched_name", "nba_sticky"]].drop_duplicates(),
    on=["matched_name"], how="left"
)
player_level["nba_sticky"] = player_level["nba_sticky"].fillna(0).astype(int)

# Features/labels
X = player_level.drop(columns=["matched_name", "nba_sticky"])
y = player_level["nba_sticky"]

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X.fillna(0), y, test_size=0.25, random_state=42
)

from sklearn.linear_model import LogisticRegression
logit = LogisticRegression(max_iter=2000, solver="lbfgs")
logit.fit(X_train, y_train)
y_pred = logit.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_pred)
print(f"AUC (Logit): {auc:.3f}")

coef_df = pd.DataFrame({
    "feature": X.columns,
    "coef": logit.coef_[0]
}).sort_values("coef", ascending=False)
print(coef_df)

```
